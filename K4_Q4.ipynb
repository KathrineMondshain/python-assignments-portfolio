```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point):

    #Function that approximates the derivative of a function at a point
    def deriv(f, base_point):
        return (f(base_point+10**(-10))-f(base_point-10**(-10)))/(2*10**(-10)) 
    
    #Sets initial x,y-coordinate
    x_coords = [initial_point]
    y_coords = [f(initial_point)]
   
    #While loop to continue adding to the lists of x and y coordinates until either a minimum is found or it has iterated 10,000 times without finding a minimum
    n = 1 #n is the number of iterations 
    while abs(deriv(f, x_coords[n-1]))>10**(-10) and n<10000:
        x_coords.append(x_coords[n-1] - learning_rate*deriv(f,x_coords[n-1])) #Appends next x-value using the previous one
        y_coords.append(f(x_coords[n])) #Appends the corresponding y-value
        n+=1
        if(n==10000): #Lets us know if we reached the cutoff, this means we could not find the minimum
            print("uhoh")
  

    #plotting section
    plot_range = np.linspace(min(x_coords)-0.5, max(x_coords)+0.5, 10000)
    function_range=[f(i) for i in plot_range]
    plt.plot(plot_range, function_range) #plots f(x)
    plt.plot(x_coords, y_coords) #plots sequence of points x_n, f(x_n)

    print(round(x_coords[-1], 3), ", ", round(y_coords[-1], 3)) #prints the last x_n and y_n rounded to 3 decimal places

#Test functions:
def f1(x):
    return x**2

def f2(x):
    return x**4 - 2*x**2

def funny_function(x):
    if x>0:
        return x**x
    elif x==0:
        return 1
    else:
        return abs(x)**abs(x)
    
def f4(x):
    return abs(x)

#gradient_descent(f1,0.9,5)

#gradient_descent(f2,0.1,0.5)
#gradient_descent(f2,0.1,-0.5)
#gradient_descent(f2,0.1,0) #Doesn't draw anything b/c f'(0) = 0, so the while loop doesn't run

#gradient_descent(funny_function, 0.1, 0.1)
#gradient_descent(funny_function, 0.1, -0.1)

gradient_descent(f4, 0.1, 1) #Doesn't work because there is no x, where f'(x) = 0. The absolute min is at x=0 but f'(0) is undefined
```
